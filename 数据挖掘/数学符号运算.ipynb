# 求极限

```python
#当x趋于0时的极限
print(limit(sin(x)/x, x, 0))

#当x趋于0时的左极限
print(limit(sin(x)/x, x, 0,"-"))

#当x趋于+oo时的极限
print(limit(sin(x)/x, x, +oo))
```

# 解非线性方程组：
非线性方程，就是因变量与自变量之间的关系不是线性的关系，这类方程很多，例如平方关系、对数关系、指数关系、三角函数关系等等  
sympy对非线性方程的求解速度很慢，有可能解不出来，scipy的求解速度比sympy快300倍

```python
solve([exp(x) - sin(y), 1/y - 3], [x, y])
```

# scipy解非线性方程求解f(x)=2*sin(x)-x+1

```python
from scipy.optimize import root,fsolve
import numpy as np
import math

def func(x):
    return x**3+x**2-1000

#第一个参数为我们需要求解的方程，第二个参数为方程解的估计值,只是影响计算速度，可随意填写
fsolve(func,[1])
```

# 解微分方程：

```python
f, g = symbols('f g', cls=Function)
dsolve(Eq(f(x).diff(x, x) - 2*f(x).diff(x) + f(x), sin(x)), f(x))
```
